##  **2. Performing Customers/sellers Segmentation**

### **a. Divide the customers into groups based on the revenue generated**

```python
merged_df.columns
```

```python
customer_revenue = merged_df.groupby('customer_unique_id')['payment_value'].sum()
bins = pd.qcut(customer_revenue, q=3, labels=['Low Revenue', 'Medium Revenue', 'High Revenue'])

customer_groups = bins.reset_index()
customer_groups.columns = ['customer_unique_id', 'revenue_group']

customer_groups
```

### **b. Divide the sellers into groups based on the revenue generated** 

```python
seller_revenue = merged_df.groupby('seller_id')['payment_value'].sum()

bins = pd.qcut(seller_revenue, q=3, labels=['Low Revenue', 'Medium Revenue', 'High Revenue'])

seller_groups = bins.reset_index()
seller_groups.columns = ['seller_id', 'revenue_group']

seller_groups
```

### **3. Cross-Selling (Which products are selling together)** 
### **Hint: We need to find which of the top 10 combinations of products are selling together in each transaction.  (combination of 2 or 3 buying together)** 

```python
combined_data = pd.merge(orders, order_items, how='left', on='order_id')

cross_sold_products = combined_data[combined_data.duplicated(['order_purchase_timestamp','customer_id'],
                                                              keep=False)].sort_values(by='order_purchase_timestamp')
cross_sold_products = cross_sold_products.loc[:,['customer_id','order_purchase_timestamp','product_id']]

cross_sold_products.drop_duplicates(subset="product_id", inplace=True)
cross_sold_products = cross_sold_products[cross_sold_products.duplicated(['order_purchase_timestamp','customer_id'],
                                                                          keep=False)].sort_values(by='order_purchase_timestamp').reset_index()
cross_sold_products.drop(columns='index', axis=1, inplace=True)

print('Products with same customer id and order purchase timestamp are cross selling')
cross_sold_products
```
